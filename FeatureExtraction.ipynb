Features with their definitions:

1. Mean
   - Definition: The average value of all points in the signal, calculated by summing all values and dividing by the number of points.

2. Standard Deviation
   - Definition: A measure of the dispersion of signal values from the mean, indicating how spread out the values are.

3. Skewness
   - Definition: Measures the asymmetry of the signal's probability distribution, indicating whether data is skewed left (negative) or right (positive).

4. Kurtosis
   - Definition: Measures the "tailedness" or "peakedness" of the signal's distribution compared to a normal distribution.

5. Root Mean Square (RMS)
   - Definition: The square root of the mean of squared values, representing the signal's overall energy or magnitude.

6. Peak-to-Peak Amplitude
   - Definition: The difference between the maximum and minimum values in the signal, showing the total signal swing.

7. Zero Crossing Rate
   - Definition: The rate at which the signal changes from positive to negative or vice versa, indicating frequency content.

8. Signal Power
   - Definition: The average of squared signal values, representing the overall strength or intensity of the signal.

9. Maximum Amplitude
   - Definition: The highest value present in the signal segment.

10. Minimum Amplitude
    - Definition: The lowest value present in the signal segment.

Frequency Domain Features:

11. Dominant Frequency
    - Definition: The frequency component with the highest amplitude in the signal's spectrum.

12. Power Spectral Density (PSD)
    - Definition: Shows how the power of the signal is distributed across different frequencies.

13. Entropy
    - Definition: A measure of the signal's randomness or unpredictability, indicating its complexity.

14. Hjorth Parameters
    - Activity: Variance of the signal (power)
    - Mobility: Ratio of the standard deviation of the first derivative to the standard deviation of the signal
    - Complexity: Ratio of the mobility of the first derivative to the mobility of the signal

Key Additions

***Hurst Exponent:*** Calculated using the hurst_exponent function.
Captures long-term memory in the time series.

***Higuchi's Fractional Dimension (HFD):*** Computed with higuchi_fd function.
Quantifies the complexity of the signal

```python
import numpy as np
from scipy.stats import skew, kurtosis
from scipy.signal import welch

# Hurst Exponent function
def hurst_exponent(ts):
    lags = range(2, 20)
    tau = [np.std(np.subtract(ts[lag:], ts[:-lag])) for lag in lags]
    poly = np.polyfit(np.log(lags), np.log(tau), 1)
    return poly[0] * 2.0

# Higuchi's Fractional Dimension function
def higuchi_fd(ts, k_max=10):
    n = len(ts)
    lk = []
    x = np.array(ts)
    for k in range(1, k_max + 1):
        lm = []
        for m in range(k):
            length = 0
            for i in range(1, int(np.floor((n - m) / k))):
                length += abs(x[m + i * k] - x[m + (i - 1) * k])
            length = (length * (n - 1)) / ((n - m) * k)
            lm.append(length)
        lk.append(np.mean(lm))
    lnk = np.log(1.0 / np.arange(1, k_max + 1))
    lmlk = np.log(lk)
    hfd = np.polyfit(lnk, lmlk, 1)[0]
    return hfd

# Feature extraction function
def feature_extraction(segment):
    # Time-domain features
    mean = np.mean(segment)
    std_dev = np.std(segment)
    rms = np.sqrt(np.mean(segment**2))
    peak_to_peak = np.ptp(segment)
    skewness = skew(segment)
    kurt = kurtosis(segment)
    max_amp = np.max(segment)
    min_amp = np.min(segment)

    # Frequency-domain features
    freqs, psd = welch(segment, fs=10000)  # 10,000 Hz sampling frequency
    dominant_freq = freqs[np.argmax(psd)]
    signal_power = np.sum(psd)

    # Hurst Exponent
    hurst = hurst_exponent(segment)

    # Higuchi's Fractional Dimension
    hfd = higuchi_fd(segment, k_max=10)

    return {
        "mean": mean,
        "std_dev": std_dev,
        "rms": rms,
        "peak_to_peak": peak_to_peak,
        "skewness": skewness,
        "kurtosis": kurt,
        "max_amp": max_amp,
        "min_amp": min_amp,
        "dominant_freq": dominant_freq,
        "signal_power": signal_power,
        "hurst_exponent": hurst,
        "higuchi_fd": hfd,
    }
```
